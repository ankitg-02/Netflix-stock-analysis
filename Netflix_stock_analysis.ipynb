```python
import numpy as np
import pandas as pd
```

```python
#importing the dataset
data=pd.read_csv("/content/NFLX (1).csv")
stock=pd.DataFrame(data)
stock.head()
```

```python
#converting the string to datetime format
stock['Date']=pd.to_datetime(stock["Date"])
```

```python
#getting the size of dataset
stock.shape
```

```python
#checking the null values of the dataset
stock.isnull().sum()
```

```python
#describing the dataset
stock.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
#relationship between date and opening price
plt.plot(stock['Date'],stock['Open'],color="black",label="Open Price")
plt.title("Opening Price over time")
plt.xlabel("Date")
plt.ylabel("Open")
plt.legend()
plt.grid(True)


```

```python
#relationship between date and closing price
plt.plot(stock['Date'],stock['Close'],color="blue",label="CLose Price")
plt.title("Closing Price over time")
plt.xlabel("Date")
plt.ylabel("Close")
plt.legend()
plt.grid(True)
```

```python
#relationship between date and the highest price
plt.plot(stock['Date'],stock['High'],color="red",label="Highest Price")
plt.title("Highest Price over time")
plt.xlabel("Date")
plt.ylabel("High")
plt.legend()
plt.grid(True)
```

```python
#relationship between date and the lowest price
plt.plot(stock['Date'],stock['Low'],color="green",label="Lowest Price")
plt.title("Lowest Price over time")
plt.xlabel("Date")
plt.ylabel("Low")
plt.legend()
plt.grid(True)
```

```python
#relationship between date and adjacent close price
plt.plot(stock['Date'],stock['Adj Close'],color="yellow",label="Adj Close Price")
plt.title("Adjacent Close Price over time")
plt.xlabel("Date")
plt.ylabel("Adj Close")
plt.legend()
plt.grid(True)
```

```python
#relationship between date and the volume
plt.plot(stock['Date'],stock['Volume'],color="brown",label="Volume")
plt.title("Volume over time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.legend()
plt.grid(True)
```
